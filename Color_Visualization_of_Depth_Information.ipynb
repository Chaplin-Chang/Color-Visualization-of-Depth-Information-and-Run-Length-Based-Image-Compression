```python
import numpy as np
import cv2
import struct
from matplotlib import pyplot as plt
rawfile = np.fromfile('02.raw', dtype=np.float32)# 以float32讀圖片
imgL = cv2.imread('02.raw',0)
imgR = cv2.imread('02.raw',0)
width = struct.unpack('i', rawfile[1])[0]
height = struct.unpack('i', rawfile[2])[0]
#disparity = stereo.compute(imgL,imgR)

PIX_ARR=[]
for I in range(8, len(rawfile)):
    pixel = struct.unpack('f', rawfile[I])[0]
    if pixel == -99999:
        pixel = np.nan
    PIX_ARR.append(pixel)

PIX_ARR = np.array(PIX_ARR)
minN = np.nanmin(PIX_ARR)
maxN = np.nanmax(PIX_ARR)

for J in range(len(PIX_ARR)):
    PIX_ARR[J] = PIX_ARR[J] / (maxN - minN) * 255
    
disp_v2 = np.nan_to_num(PIX_ARR).reshape(height, width).astype(np.uint8)
disp_v2 = cv2.applyColorMap(disp_v2, cv2.COLORMAP_JET)

plt.imshow(disp_v2)

cv2.imwrite('depth_1.png', disp_v2)
plt.show()
```
